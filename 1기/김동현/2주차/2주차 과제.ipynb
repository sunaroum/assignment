```python
# Feature Selection - Mutual Information

import pandas as pd
import numpy as np
from sklearn.metrics import mutual_info_score

def calculate_mutual_info(x, y):
    return mutual_info_score(x, y)

def calculate_avg_mutual_info(df):
    n_cols = df.shape[1]
    mutual_info_matrix = np.zeros((n_cols, n_cols))

    for i in range(n_cols):
        for j in range(i, n_cols):
            mutual_info = calculate_mutual_info(df.iloc[:, i], df.iloc[:, j])
            mutual_info_matrix[i, j] = mutual_info
            mutual_info_matrix[j, i] = mutual_info

    avg_mutual_info = np.mean(mutual_info_matrix, axis=1)
    return avg_mutual_info

def sort_columns_by_avg_mutual_info(df):
    avg_mutual_info = calculate_avg_mutual_info(df)
    sorted_columns = df.columns[np.argsort(avg_mutual_info)]
    sorted_avg_mutual_info = np.sort(avg_mutual_info)
    sorted_df = df[sorted_columns]
    return sorted_df, sorted_avg_mutual_info
```

```python
sorted_df, sorted_avg_mutual_info = sort_columns_by_avg_mutual_info(df)

print("Sorted DataFrame by Avg. Mutual Information:")
print(sorted_df)

print("\nAvg. Mutual Information for each column (in ascending order):")
for col, mi in zip(sorted_df.columns, sorted_avg_mutual_info):
    print(f"{col}: {mi:.4f}")
```
